```python
import dlib
import cv2
import os
# import ntpath as path
```

```python
#crop img
def crop_img(img, f):
    if f.top() < 0 or f.bottom() < 0 or f.left() < 0 or f.right() < 0 :
        print(f)
    crop = img[abs(f.top()):abs(f.bottom()), abs(f.left()):abs(f.right())]
    return crop

```

```python
emotion_list = []

# type category
emotion_type = ['neutral', 'anger', 'surprise', 'smile', 'sad']

# keyword by category
emotion_type_dict = {
    'neutral': ['neutral'],
    'anger': ['angry', 'anger', 'unhappy', 'upset'],
    'surprise' : ['surprise', 'shocked', 'frightened', 'amazed'],
    'smile': ['smile', 'happy', 'smiling'],
    'sad': ['sad']
}

#checking keyword
def emotion_keyword_check(name):
    emotion_count = 0
    for emotion in emotion_type:
        for keyword in emotion_type_dict[emotion]:
            if keyword in name:
                emotion_list.append(emotion)
                emotion_count += 1
                
                print(emotion, name, keyword in name)
    
#     print(emotion_list)
    if emotion_count != 1:
        for i in range(emotion_count):
            del emotion_list[-1]
    
    if emotion_count == 1:
        print(emotion_list[-1], name)
        return emotion_list[-1]
    else:
        return 'undefined'
```
